```python
from clustimage import Clustimage
import random
import os
import cv2
from pathlib import Path
import shutil
```

```python
cl = Clustimage()
```

```python
path = 'Data_CMC_MEL_patch_gray_1/train/Mitosis/'
```

```python
files=os.listdir(path)
pics=random.choices(files, k=1000)
#os.mkdir('clustimage')
```

```python
os.mkdir('gray_images')
```

```python
#print(pics)
for x in pics:
    img = cv2.imread(path+x)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    cv2.imwrite('gray_images/'+x, gray)
```

```python
for x in pics:
    shutil.copy(path+x, 'clustimage')
#print(pics)
```

```python
results = cl.fit_transform('gray_images', max_clust=5)#min_clust=10)
#results = cl.fit_transform('clustimage', min_clust=10)
```
